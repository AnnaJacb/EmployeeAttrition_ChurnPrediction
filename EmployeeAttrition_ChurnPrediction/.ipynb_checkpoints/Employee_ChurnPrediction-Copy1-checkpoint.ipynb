```python
# Fit on whole dataset to include all labels in index.
labelIndexer = StringIndexer(inputCol="Attrition", outputCol="indexedAttrition").fit(attrition_df)
```

```python
indexer = StringIndexer(inputCol="BusinessTravel", outputCol="indBusinessTravel")
indexed_att_data = indexer.fit(attrition_df).transform(attrition_df)

indexer = StringIndexer(inputCol="Department", outputCol="indDepartment")
indexed_att_data = indexer.fit(indexed_att_data).transform(indexed_att_data)

indexer = StringIndexer(inputCol="EducationField", outputCol="indEducationField")
indexed_att_data = indexer.fit(indexed_att_data).transform(indexed_att_data)

indexer = StringIndexer(inputCol="Gender", outputCol="indGender")
indexed_att_data = indexer.fit(indexed_att_data).transform(indexed_att_data)

indexer = StringIndexer(inputCol="JobRole", outputCol="indJobRole")
indexed_att_data = indexer.fit(indexed_att_data).transform(indexed_att_data)

indexer = StringIndexer(inputCol="MaritalStatus", outputCol="indMaritalStatus")
indexed_att_data = indexer.fit(indexed_att_data).transform(indexed_att_data)

indexer = StringIndexer(inputCol="Over18", outputCol="indOver18")
indexed_att_data = indexer.fit(indexed_att_data).transform(indexed_att_data)

indexer = StringIndexer(inputCol="OverTime", outputCol="indOverTime")
indexed_att_data = indexer.fit(indexed_att_data).transform(indexed_att_data)

indexed_att_data.printSchema()
```

### 4.1 Encoding

When using Machine Learning algorithms, a dataset can have only numerical values as a predictor variable. Therefore Label Encoding is necessary because categorical labels are encoded with numerical values. Hence to avoid categorical features being introduced a large number of unique values, are encoded using Label Encoding and OneHotEncoder below.

```python
from pyspark.ml.feature import (VectorAssembler,VectorIndexer,
                                OneHotEncoder,StringIndexer)
```

```python
# First create a string indexer (convert every string into a number, such as male = 0 and female = 1).
# A number will be assigned to every category in the column.
attrition_indexer = StringIndexer(inputCol='Attrition',outputCol='AttrtionIndex').fit(attrition_df_target)

# Now we can one hot encode these numbers. This converts the various outputs into a single vector.
# This makes it easier to process when you have multiple classes.
attrition_encoder = OneHotEncoder(inputCol='AttrtionIndex',outputCol='AttritionVector')
```

### 4.2 Spark Formatting of Data to model  

Splitting data into training and testing sets before implementing Machine Learning algorithms.

```python
# A few things we need to do before Spark can accept the data!
# It needs to be in the form of two columns: "label" and "features".

# Import VectorAssembler and Vectors
from pyspark.ml.linalg import Vectors
from pyspark.ml.feature import VectorAssembler
data = spark.read.csv(path,inferSchema=True,header=True)
```

```python
# The input columns are the feature column names, and the output column is what you'd like the new column to be named. 
assembler = VectorAssembler(
    inputCols=["Attrition_numerical","DailyRate","DistanceFromHome","Education","EmployeeNumber","EnvironmentSatisfaction"],
    outputCol="features")
```

```python
# Let's transform the data. 
output = assembler.transform(attrition_df_target_small)
```

```python
# Let's import the string indexer (similar to the logistic regression exercises).
from pyspark.ml.feature import StringIndexer
```

```python
indexer = StringIndexer(inputCol="Attrition_numerical", outputCol="AttritionIndex")
output_fixed = indexer.fit(output).transform(output)
```

```python
# Let's select the two columns we want. Features (which contains vectors), and the predictor.
final_data = output_fixed.select("features",'AttritionIndex')
```

```python
# Split the training and testing set.
training_data,testing_data = final_data.randomSplit([0.7,0.3])
```

## 5 Implementing Machine Learning Models

### Logistic Regression

```python
# Let's import the relevant classifiers. 
from pyspark.ml.classification import LogisticRegression
from pyspark.ml import Pipeline
```

```python
# Note that attrition is a categorial variable but didn't require any transformation.
# That's because it's already in the format of 1's and 0's. 
log_attrition = LogisticRegression(featuresCol='features',labelCol='AttritionIndex')
```

```python
# Lists everything we want to do. Index data, encode data, assemble data and then pass in the actual model.
# attrition_encoder
pipeline = Pipeline(stages=[log_attrition, labelIndexer, attrition_indexer])
```

```python
# Note pipeline. Call it as you would call a machine learning object.
fit_model = pipeline.fit(training_data)
```

```python
# Transform test data. 
results = fit_model.transform(testing_data)
```

```python
# Evaluate the model using the binary classifer.
from pyspark.ml.evaluation import BinaryClassificationEvaluator

my_eval = BinaryClassificationEvaluator(rawPredictionCol='prediction',
                                       labelCol='AttritionIndex')
```

```python
# If we select the actual and predicted results, we can see that some predictions were correct while others were wrong.
results.select('AttritionIndex','prediction')
```

```python
# We can then evaluate using AUC (area under the curve). AUC is linked to ROC.
AUC = my_eval.evaluate(results)

AUC
```

### The tree Classifiers

```python
# Let's import the relevant classifiers. 
from pyspark.ml.classification import DecisionTreeClassifier,GBTClassifier,RandomForestClassifier
from pyspark.ml import Pipeline
```

Create all three models:

```python
# Use defaults to make the comparison "fair". This simplifies the comparison process.

dtc = DecisionTreeClassifier(labelCol='AttritionIndex',featuresCol='features')
rfc = RandomForestClassifier(labelCol='AttritionIndex',featuresCol='features')
gbt = GBTClassifier(labelCol='AttritionIndex',featuresCol='features')
```

Train all three models:

```python
# Train the models (it's three models, so it might take some time).
dtc_model = dtc.fit(training_data)
rfc_model = rfc.fit(training_data)
gbt_model = gbt.fit(training_data)
```

## Model Comparison

Let's compare each of these models!

```python
dtc_predictions = dtc_model.transform(testing_data)
rfc_predictions = rfc_model.transform(testing_data)
gbt_predictions = gbt_model.transform(testing_data)
```

**Evaluation Metrics:**

```python
# Let's start off with binary classification.
from pyspark.ml.evaluation import BinaryClassificationEvaluator

# Note that the label column isn't named label, it's named PrivateIndex in this case.
my_binary_eval = BinaryClassificationEvaluator(labelCol = 'AttritionIndex')
```

```python
# This is the area under the curve. This indicates that the data is highly seperable.
print("DTC")
print(my_binary_eval.evaluate(dtc_predictions))

# RFC improves accuracy but also model complexity. RFC outperforms DTC in nearly every situation.
print("RFC")
print(my_binary_eval.evaluate(rfc_predictions))

# We can't repeat these exact steps for GBT. If you print the schema of all three, you may be able to notice why.
my_binary_gbt_eval = BinaryClassificationEvaluator(labelCol='AttritionIndex', rawPredictionCol='prediction')
print("GBT")
print(my_binary_gbt_eval.evaluate(gbt_predictions))
```

```python
# Let's import the evaluator.
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
# Select (prediction, true label) and compute test error. 
acc_evaluator = MulticlassClassificationEvaluator(labelCol="AttritionIndex", predictionCol="prediction", metricName="accuracy")
```

```python
dtc_acc = acc_evaluator.evaluate(dtc_predictions)
rfc_acc = acc_evaluator.evaluate(rfc_predictions)
gbt_acc = acc_evaluator.evaluate(gbt_predictions)
```

```python
# Let's do something a bit more complex in terms of printing, just so it's formatted nicer. 
print("Here are the results!")
print('-'*40)
print('A single decision tree has an accuracy of: {0:2.2f}%'.format(dtc_acc*100))
print('-'*40)
print('A random forest ensemble has an accuracy of: {0:2.2f}%'.format(rfc_acc*100))
print('-'*40)
print('An ensemble using GBT has an accuracy of: {0:2.2f}%'.format(gbt_acc*100))
```
